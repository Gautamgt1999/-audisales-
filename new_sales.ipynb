```python

from sqlalchemy import text


setup_queries = """
-- Create Make table
CREATE TABLE IF NOT EXISTS make (
    make_id SERIAL PRIMARY KEY,
    make_name VARCHAR(100) UNIQUE NOT NULL
);

-- Populate Make table
INSERT INTO make (make_name)
SELECT DISTINCT "Make" FROM carsale1
WHERE "Make" IS NOT NULL
ON CONFLICT (make_name) DO NOTHING;

-- Create FuelType table
CREATE TABLE IF NOT EXISTS fuel_type (
    fuel_type_id SERIAL PRIMARY KEY,
    fuel_type_name VARCHAR(50) UNIQUE NOT NULL
);

-- Populate FuelType table
INSERT INTO fuel_type (fuel_type_name)
SELECT DISTINCT "fuelType" FROM carsale1
WHERE "fuelType" IS NOT NULL
ON CONFLICT (fuel_type_name) DO NOTHING;

-- Create Transmission table
CREATE TABLE IF NOT EXISTS transmission (
    transmission_id SERIAL PRIMARY KEY,
    transmission_name VARCHAR(50) UNIQUE NOT NULL
);

-- Populate Transmission table
INSERT INTO transmission (transmission_name)
SELECT DISTINCT "transmission" FROM carsale1
WHERE "transmission" IS NOT NULL
ON CONFLICT (transmission_name) DO NOTHING;
"""

# Execute setup
with engine.connect() as conn:
    conn.autocommit = True
    for query in setup_queries.split(';'):
        if query.strip():
            try:
                conn.execute(text(query))
                print(f" Executed: {query[:50]}...")
            except Exception as e:
                print(f"! Error: {e}")

print("\n Relationships and lookup tables created successfully!")
```

```python
# Diagnostic: check PostgreSQL connection and existence of 'carsale1' table
from sqlalchemy import inspect
import pandas as pd

print('Engine URL:', getattr(engine, 'url', engine))
inspector = inspect(engine)
tables = inspector.get_table_names()
print('Existing tables in database:', tables)

if 'carsale1' not in tables:
    print("'carsale1' not found  attempting to upload dataframe to create the table...")
    try:
        df.to_sql('carsale1', engine, if_exists='replace', index=False, method='multi', chunksize=1000)
        print("Upload successful: 'carsale1' created.")
    except Exception as e:
        print('Upload failed:', e)
else:
    print("'carsale1' already exists  no action taken.")


print('Tables after check:', inspector.get_table_names())

# Optional quick preview if table now exists
if 'carsale1' in inspector.get_table_names():
    preview = pd.read_sql('SELECT * FROM carsale1 LIMIT 5', engine)
    print('\nPreview of carsale1:')
    print(preview)
```

```python

from sqlalchemy import text


query = "SELECT * FROM carsale1 LIMIT 10"
result_df = pd.read_sql(query, engine)
print("First 10 rows from carsale1:")
print(result_df)


query = """
SELECT "Make", COUNT(*) as count, AVG("price") as avg_price 
FROM carsale1 
GROUP BY "Make" 
ORDER BY count DESC
"""
summary_df = pd.read_sql(query, engine)
print("\n\nSummary by Make:")
print(summary_df)


query = """
SELECT "fuelType", COUNT(*) as count, AVG("price") as avg_price 
FROM carsale1 
GROUP BY "fuelType" 
ORDER BY count DESC
"""
fuel_df = pd.read_sql(query, engine)
print("\n\nSummary by FuelType:")
print(fuel_df)


query = """
SELECT "transmission", COUNT(*) as count, AVG("price") as avg_price 
FROM carsale1 
GROUP BY "transmission" 
ORDER BY count DESC
"""
trans_df = pd.read_sql(query, engine)
print("\n\nSummary by Transmission:")
print(trans_df)
```
